# Write a python program to merge data from each month into one CSV and read in updated dataframe. 

```python
#importing all the needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import seaborn
```

```python
#getting all the csv files in the current directory 
files = glob.glob("Sales_*.csv")
files
```

```python
#creating an empty list to store the dataframe
arrayy = []

#looping through each file and reading into a dataframe
for file in files:
    df = pd.read_csv(file)
    arrayy.append(df)

#concatenating all dataframes inside the list into one dataframe
sales = pd.concat(arrayy)
sales
```

```python
#saving the merged dataframe into a new CSV file
sales.to_csv("Sales.csv", encoding='utf-8', index=False)

#reading the updated CSV as a dataframe
newSale = pd.read_csv("Sales.csv")
newSale
```

# Which Month has the best sales? and how much was the earning in that month? Make a bar graph of sales as well. 

```python
#creating a new column named Sales that is product of Quantity ordered and price each
newSale["Sales"] = newSale["Quantity Ordered"] * newSale["Price Each"]

#checking in the dataframe if the new column is formed correctly
newSale
```

```python
#grouping the data by month and calulating the sum of sales
total_sales = newSale.groupby(["Month"])["Sales"].sum().reset_index()

#finding the month with the highest sales
total_sales[total_sales["Sales"] == total_sales["Sales"].max()]
```

```python

total_sales
```

```python
#plotting a bar chart of product sales, with months names on the x-axis and sales on y-axis
total_sales.plot.bar(x= "Month", y = "Sales", rot= 0, title = "Sales in Month", figsize= (18,8), color = 'lavender', edgecolor= "purple")

#adding labels to the x and y axis
plt.xlabel("Months")
plt.ylabel("Sales")

#setting the x-tick labels to the months names
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.xticks(total_sales.index, months)

#Setting the y-axis limits to start from 1700000 and end at 5000000 for better visualization
plt.ylim(1700000, 5000000)

#Adding the value of the Sales sold for each month above the respective bars
for i, j in enumerate(total_sales["Sales"]):
    plt.text(i-.30, j+ 20000, str(round(j)), color= 'purple')
    
#Displaying the bar chart
plt.show()
```

# Which city has sold the highest product?

```python
#grouping the data by city and calulating the sum of all products (quantity ordered)
sales_city = newSale.groupby(["City"])["Quantity Ordered"].sum().reset_index()

#finding the city with the highest product
sales_city[sales_city["Quantity Ordered"] == sales_city["Quantity Ordered"].max()]
```

# Which product was sold the most in overall? Illustrate it through bar graph.

```python
#grouping the data by product and calulating the sum of total ordered data
sales_product = newSale.groupby(["Product"])["Quantity Ordered"].sum().reset_index()

#finding the product with the highest sales
sales_product[sales_product["Quantity Ordered"] == sales_product["Quantity Ordered"].max()]
```

```python
sales_product
```

```python
#plotting a bar chart of product sales, with product names on the x-axis and quantity sold on y-axis
sales_product.plot.bar(x= "Product", y = "Quantity Ordered", rot= 90, title = "Product's overall Sale", figsize= (20,8), color = 'lavender', edgecolor = "purple")

#adding labels to the x and y axis
plt.xlabel("Product")
plt.ylabel("Quantity Sold")

#setting the x-tick labels to the product names which are rotated to 90 degree for better readability
plt.xticks(sales_product.index, sales_product["Product"])

#Setting the y-axis limits to start from 200 and end at 32000 for better visualization
plt.ylim(200, 32000)

#Adding the value of the quantity sold for each product above the respective bars
for i, j in enumerate(sales_product["Quantity Ordered"]):
    plt.text(i-.27, j+ 80, str(round(j)), color= 'purple')
    
#Displaying the bar chart
plt.show()
```
